#4.4. Spanning Sets and Linear Independence

####$\star$Example
Now, we are interested in something like:<br>
Given $$c_1\langle 2, 1 \rangle+c_2\langle-1, 1 \rangle=c_1\begin{bmatrix}2 \\ 1 \end{bmatrix}+c_2\begin{bmatrix}-1 \\ 1 \end{bmatrix}=\begin{bmatrix}0 \\ -2 \end{bmatrix},$$
what are $c_1$ and $c_2$?


```r
# Vector u
u_1 = 2
u_2 = 1
# Vector v
v_1 = -1
v_2 = 1
# Vector b
b_1 = 0
b_2 = -2

axis <- c(-5,5)
plot(axis,axis)
grid()
arrows(0,0,u_1,u_2,lwd=4,col="red")
arrows(0,0,v_1,v_2,lwd=4,col="blue")
arrows(0,0,b_1,b_2,lwd=4,col="black")
text(1.8,1.5,"u", cex=2)
text(-0.8,1.5,"v", cex=2)
text(0.8,-1.5,"b", cex=2)
```

To solve for $c_1$ and $c_2$, we write it in a matrix form:
$$c_1\begin{bmatrix}2 \\ 1 \end{bmatrix}+c_2\begin{bmatrix}-1 \\ 1 \end{bmatrix}=\begin{bmatrix}2 &-1\\ 1 &1\end{bmatrix}\begin{bmatrix}c_1 \\ c_2 \end{bmatrix}=\begin{bmatrix}0 \\ -2 \end{bmatrix},$$
that is,
\begin{align}
\begin{bmatrix}c_1 \\ c_2 \end{bmatrix}&=\begin{bmatrix}2 &-1\\ 1 &1\end{bmatrix}^{-1}\begin{bmatrix}0 \\ -2 \end{bmatrix}\\
&=\begin{bmatrix}\frac{1}{3} &\frac{1}{3}\\ -\frac{1}{3} &\frac{2}{3}\end{bmatrix}\begin{bmatrix}0 \\ -2 \end{bmatrix}\\
&=\begin{bmatrix}-\frac{2}{3}  \\ -\frac{4}{3}  \end{bmatrix}
\end{align}
So, $c_1=-\frac{2}{3}, c_2=-\frac{4}{3}$ satisfies the equation 
$$-\frac{2}{3}\begin{bmatrix}2 \\ 1 \end{bmatrix}-\frac{4}{3}\begin{bmatrix}-1 \\ 1 \end{bmatrix}=\begin{bmatrix}-\frac{2}{3}(2)--\frac{4}{3}\\-\frac{2}{3}(1)-\frac{4}{3}(1) \end{bmatrix}=\begin{bmatrix}0 \\ -2 \end{bmatrix}$$
<br>
That is, a vector $\begin{bmatrix}0 \\ -2 \end{bmatrix}$ can be expressed with two vectors $\begin{bmatrix}2 \\ 1 \end{bmatrix}$, $\begin{bmatrix}-1 \\ 1 \end{bmatrix}$ and coefficients.
$$-\frac{2}{3}\begin{bmatrix}2 \\ 1 \end{bmatrix}-\frac{4}{3}\begin{bmatrix}-1 \\ 1 \end{bmatrix}=\begin{bmatrix}0 \\ -2 \end{bmatrix}$$


```r
# Vector u
u_1 = 2
u_2 = 1
# Vector v
v_1 = -1
v_2 = 1
# Vector w
b_1 = 0
b_2 = -2

# Coefficients
c_1 = -2/3
c_2 = -4/3

axis <- c(-5,5)
plot(axis,axis)
grid()
arrows(0,0,u_1,u_2,lwd=4,col="red")
arrows(0,0,v_1,v_2,lwd=4,col="blue")
arrows(0,0,b_1,b_2,lwd=4,col="black")

arrows(0,0,c_1*u_1,c_1*u_2,lwd=2,col="red",lty = "dashed")
arrows(0,0,c_2*v_1,c_2*v_2,lwd=2,col="blue",lty = "dashed")
arrows(c_1*u_1,c_1*u_2,c_1*u_1+c_2*v_1,c_1*u_2+c_2*v_2,lwd=2,col="blue",lty = "dashed")
arrows(0,0,c_1*u_1+c_2*v_1,c_1*u_2+c_2*v_2,lwd=4)

arrows(0,0,b_1,b_2,lwd=4,col="black")
text(1.8,1.5,"u", cex=2)
text(-2,-0.2,"-2/3u", cex=2)
text(-0.8,1.5,"v", cex=2)
text(-1.2,-2,"-4/3v", cex=2)
text(0.8,-1.5,"b", cex=2)
```

##Definition of a Linear Combination of Vectors
A vector $\textbf{v}$ in a vector space $V$ is a **linear combination** of the vectors $\textbf{u}_1,\textbf{u}_2,\ldots,\textbf{u}_k$ in $V$ when $\textbf{v}$ can be written in the form
$$\textbf{v}=c_1\textbf{u}_1+c_2\textbf{u}_2+\cdots+c_k\textbf{u}_k$$
where $c_1,c_2,\ldots,c_k$ are scalars.


####$\star$Example
(a) For the set of vectors in $\mathbb{R}^3$
$$S=\{\textbf{v}_1,\textbf{v}_2,\textbf{v}_3\}=\{\langle1,3,1\rangle,\langle0,1,2\rangle,\langle1,0,-5\rangle\}$$ 
$\textbf{v}_1$ is a linear combination of $\textbf{v}_2$ and $\textbf{v}_3$ because
$$\textbf{v}_1=3\textbf{v}_2+\textbf{v}_3=3\langle0,1,2\rangle+\langle1,0,-5\rangle=\langle1,3,1\rangle$$

(b) For the set of vectors in $M_{2,2}$
$$S=\{\textbf{v}_1,\textbf{v}_2,\textbf{v}_3,\textbf{v}_4\}=\left\{\begin{bmatrix}0&8\\2&1\end{bmatrix},\begin{bmatrix}0&2\\1&0\end{bmatrix},\begin{bmatrix}-1&3\\1&2\end{bmatrix},\begin{bmatrix}-2&0\\1&3\end{bmatrix}\right\}$$ 
$\textbf{v}_1$ is a linear combination of $\textbf{v}_2$, $\textbf{v}_3$ and $\textbf{v}_4$ because
\begin{align*}
\textbf{v}_1&=\textbf{v}_2+2\textbf{v}_3-\textbf{v}_4\\
&=\begin{bmatrix}0&2\\1&0\end{bmatrix}+2\begin{bmatrix}-1&3\\1&2\end{bmatrix}-\begin{bmatrix}-2&0\\1&3\end{bmatrix}\\
&=\begin{bmatrix}0&8\\2&1\end{bmatrix}.
\end{align*}

```r
# Example
# Write the vector w=<1,1,1> as a linear combination of vectors in the set
# S={<1,2,3>,<0,1,2>,<-1,0,1>}
# <1,1,1> = c1<1,2,3>+c2<0,1,2>+c3<-1,0,1>

A <- matrix(c(1,0,-1,1,2,1,0,1,3,2,1,1), 3, 4, byrow = TRUE)
rref(A)

# c1 = 1+t
# c2 = -1-2t
# c3 = t
# Let t = 1
# w = 2v_1-3v_2+v_3

B <- matrix(c(1,0,-1,1,2,1,0,-2,3,2,1,2), 3, 4, byrow = TRUE)
rref(B)

# No solution -> cannot be written as a linear combination
```

##Definition of a Spanning Set of a Vector Space
Let $S=\{\textbf{v}_1,\textbf{v}_2,\ldots,\textbf{v}_k\}$ be a subset of a vector space $V$. The set $S$ is a **spanning set** of $V$ when every vector in $V$ can be written as a linear combination of vectors in $S$. In such cases, it is said that $S$ **spans** $V$.

```r
# Vector u
u_1 = 2
u_2 = 1
# Vector v
v_1 = -1
v_2 = 1

axis <- c(-3,3)
plot(axis,axis)
grid()
arrows(0,0,u_1,u_2,lwd=4,col="red")
arrows(0,0,v_1,v_2,lwd=4,col="blue")

arrows(0,0,0,1,lwd=4,col="black",lty="dotted")
arrows(0,0,1,0,lwd=4,col="black",lty="dotted")
```

For $\textbf{v}_1=\begin{bmatrix}2\\1\end{bmatrix}$ and $\textbf{v}_2=\begin{bmatrix}-1\\1\end{bmatrix}$,<br>
<br>
$S=\{\textbf{v}_1,\textbf{v}_2\}$ spans $\mathbb{R}^2$.<br>
<br>
In other words,
$\text{Span}\{\textbf{v}_1,\textbf{v}_2\}=\mathbb{R}^2$.

```r
# Vector u
u_1 = 2
u_2 = 1
# Vector v
v_1 = 1
v_2 = 0.5

axis <- c(-3,3)
plot(axis,axis)
grid()
arrows(0,0,u_1,u_2,lwd=4,col="red")
arrows(0,0,v_1,v_2,lwd=4,col="blue")
abline(0,1/2,lty="dashed")
```

For $\textbf{v}_1=\begin{bmatrix}2\\1\end{bmatrix}$ and $\textbf{v}_2=\begin{bmatrix}1\\0.5\end{bmatrix}$,<br>
<br>
$S=\{\textbf{v}_1,\textbf{v}_2\}$ spans $\{t\langle 2,1\rangle|t\in \mathbb{R}\}$.<br>
<br>
In other words,
$\text{Span}\{\textbf{v}_1,\textbf{v}_2\}=\text{Span}\{\textbf{v}_1\}=\text{Span}\{\textbf{v}_2\}$.

##Definition of the Span of a Set
If $S=\{\textbf{v}_1,\textbf{v}_2,\ldots,\textbf{v}_k\}$ is a set of vectors in a vector space $V$, then the \textbf{span of S} is the set of all linear combinations of the vectors in $S$,
$$\text{span}(S)=\{c_1\textbf{v}_1+c_2\textbf{v}_2+\cdots+c_k\textbf{v}_k: c_1,c_2,\ldots,c_k \text{ are real numbers}\}.$$
The span of $S$ is denoted by
$$\text{span}(S)$$
or
$$\text{span}\{\textbf{v}_1,\textbf{v}_2,\ldots,\textbf{v}_k\}.$$
When $\text{span}(S)=V$, it is said that $V$ is **spanned** by $\{\textbf{v}_1,\textbf{v}_2,\ldots,\textbf{v}_k\}$, or that $S$ **spans** $V$.

###Theorem 4.7: $\text{span}(S)$ Is a Subspace of $V$
If $S=\{\textbf{v}_1,\textbf{v}_2,\ldots,\textbf{v}_k\}$ is a set of vectors in a vector space $V$, then  $\text{span}(S)$ is a subspace of $V$. Moreover,  $\text{span}(S)$ is the smallest subspace of $V$ that contains $S$, in the sense that every other subspace of $V$ that contains $S$ must contain  $\text{span}(S)$.


##Definition of Linear Dependence and Linear Independence
A set of vectors $S=\{\textbf{v}_1,\textbf{v}_2,\ldots,\textbf{v}_k\}$ in a vector space $V$ is **linearly independent** when the vector equation
$$c_1\textbf{v}_1+c_2\textbf{v}_2+\cdots+c_k\textbf{v}_k=\textbf{0}$$
has only the trivial solution
$$c_1=0,c_2=0,\ldots,c_k=0.$$
If there are also nontrivial solutions, then  $S$ is **linearly dependent**.


###Testing for Linear Independence and Linear Dependence
Let $S=\{\textbf{v}_1,\textbf{v}_2,\ldots,\textbf{v}_k\}$ be a set of vectors in a vector space $V$. To determine whether $S$ is linearly independent or linearly dependent, use the steps below.
1. From the vector equation $c_1\textbf{v}_1+c_2\textbf{v}_2+\cdots+c_k\textbf{v}_k=\textbf{0}$, write a system of linear equations in the variables $c_1=0,c_2=0$ and $c_k=0$.
2. Determine whether the system has a unique solution.
3. If the system has only the trivial solution, $c_1=0,c_2=0,\ldots,c_k=0$, then the set $S$ is linearly independent. If the system also has nontrivial solutions, then $S$ is linearly dependent.

####$\star$Example
Determine whether the set of vectors in $\mathbb{R}^3$ is linearly independent or linearly dependent.
$$S=\{\textbf{v}_1,\textbf{v}_2,\textbf{v}_3\}=\{\langle1,2,3\rangle,\langle0,1,2\rangle,\langle-2,0,1\rangle\}$$ 
<br>
**Solution**
To test for linear independence or linear dependence, form the vector equation
$$c_1\textbf{v}_1+c_2\textbf{v}_2+c_3\textbf{v}_3=0.$$
If the only solution of this equation is $c_1=c_2=c_3=0$, then the set $S$ is linearly independent. Otherwise,  is linearly dependent.<br>


\begin{alignat*}{3}
    c_1& {}{} &  & {}-{} & 2c_3 &= 0\\
    2c_1 & {}+{} & c_2 & {}{} &  &= 0\\
    3c_1& {}+{} & 2c_2 & {}+{} &c_3 &= 0\\
\end{alignat*} 



```r
A <- matrix(c(1,0,-2,0,2,1,0,0,3,2,1,0), 3, 4, byrow = TRUE)
rref(A)
```

This implies that the only solution is the trivial solution $c_1=c_2=c_3=0$. So, $S$ is linearly independent.

###Theorem 4.8: A Property of Linearly Dependent Sets
A set $S=\{\textbf{v}_1,\textbf{v}_2,\ldots,\textbf{v}_k\}, k\geq 2$, is linearly dependent if and only if at least one of the vectors $\textbf{v}_i$ can be written as a linear combination of the other vectors in $S$.

###Theorem 4.8: Corollary
Two vectors $\textbf{u}$ and $\textbf{v}$ in a vector space $V$ are linearly dependent if and only if one is a scalar multiple of the other.
